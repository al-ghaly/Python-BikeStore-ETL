```python
import pandas as pd
```

```python
import json, requests, os
results = requests.get('https://openexchangerates.org/api/latest.json',
            params={"app_id": "a92f8bad8e044bc79949a676886da2c8"})
data = results.json()
```

```python
data
```

```python
data['rates']
```

```python
extracted_data = {'Value': data['rates']['EGP'],
                  'Timestamp': data['timestamp'],
                    'Sourc': 'API'}
```

```python
df = pd.json_normalize(extracted_data)
df['Timestamp'] = pd.to_datetime(df['Timestamp'], unit='s')
```

```python
file_name = '../bikestore_dim/exchange.csv'
# Check if the file exists
file_exists = os.path.isfile(file_name)

df.to_csv(file_name, mode='a', header=not file_exists, index=False)
```
